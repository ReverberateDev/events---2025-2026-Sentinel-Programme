```python
number = int(input("Enter a number: "))

def isprime(n):
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
print(f"Prime numbers between 0 and {number} :")
for i in range(2, number + 1):
    if isprime(i):
        print(i)
```

```python
lst = []

def merge(lst, lptr, mid, rptr):
    left = lst[lptr:mid + 1]
    right = lst[mid + 1:rptr + 1]
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            lst[lptr] = left[i]
            i += 1
        else:
            lst[lptr] = right[j]
            j += 1
        lptr += 1
    while i < len(left):
        lst[lptr] = left[i]
        i += 1
        lptr += 1
    while j < len(right):
        lst[lptr] = right[j]
        j += 1
        lptr += 1
def merge_sort(lst, lptr, rptr):
    if lptr == rptr:
        return
    mid = (lptr + rptr) // 2
    merge_sort(lst, lptr, mid)
    merge_sort(lst, mid + 1, rptr)
    merge(lst, lptr, mid, rptr)

while True:
    number = input("Enter a number (enter 0 to stop): ")
    if not number.isnumeric() or int(number) == 0:
        break
    lst.append(int(number))

merge_sort(lst, 0, len(lst) - 1)
print(lst)
```

```python
import random

random_number = random.randint(1, 100)

modes = {
    "easy" : float('inf'),
    "medium" : 50,
    "hard" : 10
}

mode = input("Select game mode (easy, medium, hard): ").strip().lower()
if mode not in modes:
    print("Invalid mode selected")
else:
    tries = modes[mode]
    while tries > 0:
        guess = input(f"Guess the number between 1 and 100: ")
        if not guess.isnumeric():
            print("Please enter a valid number.")
            continue
        guess = int(guess)
        if guess == random_number:
            print(f"Congratulations! The secret number was {random_number}.")
            break
        elif guess < random_number:
            print("Too low. Try again.")
        else:
            print("Too high. Try again.")
        tries -= 1
    else:
        print("Out of tries. Game over.")

```

```python
length = int(input("Enter the length of the sequence: "))
height = int(input("Enter the height of the zigzag pattern: "))

grid = [[" " for _ in range(length)] for _ in range(height)]
dr = 1
row = 0
now = 1
for col in range(length):
    grid[row][col] = now
    now += 1
    if dr == 1 and row == height - 1:
        dr = -1
    elif dr == -1 and row == 0:
        dr = 1
    row += dr

# display the zigzag pattern
for r in range(height):
    for c in range(length):
        print(grid[r][c], end=" ")
    print()
```

```python
def display_menu():
    print("""Simple Banking System
1. Create an Account
2. Deposit Money
3. Withdraw Money
4. Check Balance
5. Exit""")

class BankAccount:
    def __init__(self, account_number, account_holder):
        self.account_number = account_number
        self.account_holder = account_holder
        self.balance = 0.0

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited: ${amount}. New balance: ${self.balance}")

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew: ${amount}. New balance: ${self.balance}")
        else:
            print("You do not have enough money.")

    def checkBalance(self):
        print(f"Current balance: {self.balance}")

accounts = []

while True:
    display_menu()
    option = input("Select an option: ")
    if option not in ['1', '2', '3', '4', '5']:
        print("Invalid option. Please choose again.")
        continue
    if option == '1':
        account_number = input("Enter account number: ")
        account_holder = input("Enter account holder name: ")
        accounts.append(BankAccount(account_number, account_holder))
    elif option == '2':
        account_number = input("Enter account number: ")
        for account in accounts:
            if account.account_number == account_number:
                amount = float(input("Enter amount to deposit: "))
                account.deposit(amount)
                break
    elif option == '3':
        account_number = input("Enter account number: ")
        for account in accounts:
            if account.account_number == account_number:
                amount = float(input("Enter amount to withdraw: "))
                account.withdraw(amount)
                break
    elif option == '4':
        account_number = input("Enter account number: ")
        for account in accounts:
            if account.account_number == account_number:
                account.checkBalance()
                break
    if option == '5':
        print("Exitiing banking system. Goodbye.")
        break
```

```python
maze = [
    ["X", "X", "X", "X", "X", "X", "X"],
    ["X", "S", " ", " ", " ", "X", "X"],
    ["X", " ", "X", "X", " ", "X", "X"],
    ["X", " ", "X", " ", " ", " ", "X"],
    ["X", " ", " ", " ", "X", "E", "X"],
    ["X", "X", "X", "X", "X", "X", "X"]
]

rows = int(len(maze))
cols = int(len(maze[0]))
directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]

for row in range(rows):
    for col in range(cols):
        if maze[row][col] == "S":
            start = (row, col)
        if maze[row][col] == "E":
            end = (row, col)


queue = [start]
visited = set(start)
parent = {start: None}

while queue:
    current = queue.pop(0)
    r, c = current

    if (r, c) == end:
        path = []
        while (r, c) != start:
            path.append((r, c))
            r, c = parent[(r, c)]
        path.append(start)
    for dy, dx in directions:
        nr, nc = r + dy, c + dx
        if 0 <= nr < rows and 0 <= nc < cols and maze[nr][nc] != "X" and (nr, nc) not in visited:
            visited.add((nr, nc))
            parent[(nr, nc)] = (r, c)
            queue.append((nr, nc))

for r, c in path[1:]:
    maze[r][c] = "*"


print("Maze before solving:")
for row in maze:
    for item in row:
        print(item, end="")
    print()

print("Maze after solving:")
for row in maze:
    for item in row:
        print(item, end="")
    print()
```

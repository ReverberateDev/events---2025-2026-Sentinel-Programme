```python
movies = ["The Shawshank Redemption", "The Godfather", "The Dark Knight", "Forrest Gump", "Inception", "The Matrix", "Avengers: Infinity War", "Back to the Future", "The Lion King", "Pulp Fiction"]

last_search = ""
last_results = []

while True:
    search_input = input("Enter movie search term(s) or 'last' to view previous search: ").strip()
    
    if search_input.lower() == "last":
        if last_search:
            print(f"Last search: {last_search}")
            for result in last_results:
                print(result)
            print(f"Found {len(last_results)} result(s).")
        else:
            print("No previous search found.")
        continue

    terms = search_input.lower().split()
    results = []

    for title in movies:
        title_lower = title.lower()
        for term in terms:
            if term in title_lower:
                results.append(title)
                break

    for r in results:
        print(r)

    print(f"Found {len(results)} result(s).")

    last_search = search_input
    last_results = results
```

```python
movies = [
    {"title": "The Shawshank Redemption", "director": "Frank Darabont", "year": 1994, "genre": "Drama"},
    {"title": "The Godfather", "director": "Francis Ford Coppola", "year": 1972, "genre": "Crime, Drama"},
    {"title": "The Dark Knight", "director": "Christopher Nolan", "year": 2008, "genre": "Action, Crime, Drama"},
    {"title": "Forrest Gump", "director": "Robert Zemeckis", "year": 1994, "genre": "Drama, Romance"},
    {"title": "Inception", "director": "Christopher Nolan", "year": 2010, "genre": "Action, Adventure, Sci-Fi"},
    {"title": "The Matrix", "director": "Lana Wachowski, Lilly Wachowski", "year": 1999, "genre": "Action, Sci-Fi"},
    {"title": "Avengers: Infinity War", "director": "Anthony Russo, Joe Russo", "year": 2018, "genre": "Action, Adventure, Sci-Fi"},
    {"title": "Back to the Future", "director": "Robert Zemeckis", "year": 1985, "genre": "Adventure, Comedy, Sci-Fi"},
    {"title": "The Lion King", "director": "Roger Allers, Rob Minkoff", "year": 1994, "genre": "Animation, Adventure, Drama"},
    {"title": "Pulp Fiction", "director": "Quentin Tarantino", "year": 1994, "genre": "Crime, Drama"},
    {"title": "Fight Club", "director": "David Fincher", "year": 1999, "genre": "Drama"},
    {"title": "Interstellar", "director": "Christopher Nolan", "year": 2014, "genre": "Adventure, Drama, Sci-Fi"},
    {"title": "Spirited Away", "director": "Hayao Miyazaki", "year": 2001, "genre": "Animation, Adventure, Family"},
    {"title": "La La Land", "director": "Damien Chazelle", "year": 2016, "genre": "Comedy, Drama, Music"},
    {"title": "Jurassic Park", "director": "Steven Spielberg", "year": 1993, "genre": "Action, Adventure, Sci-Fi"},
    {"title": "Titanic", "director": "James Cameron", "year": 1997, "genre": "Drama, Romance"},
    {"title": "The Lord of the Rings: The Fellowship of the Ring", "director": "Peter Jackson", "year": 2001, "genre": "Adventure, Drama, Fantasy"},
    {"title": "Star Wars: Episode IV - A New Hope", "director": "George Lucas", "year": 1977, "genre": "Action, Adventure, Fantasy"},
    {"title": "Goodfellas", "director": "Martin Scorsese", "year": 1990, "genre": "Biography, Crime, Drama"},
    {"title": "The Silence of the Lambs", "director": "Jonathan Demme", "year": 1991, "genre": "Crime, Drama, Thriller"},
]

def process_query(query):
    processed_query = []
    i = 0
    while i < len(query):
        if query[i] == '"':
            i += 1
            start = i
            while i < len(query) and query[i] != '"':
                i += 1
            if i < len(query):
                quoted = query[start:i]
                quoted = quoted.replace(" ", "_")
                processed_query.append(f'"{quoted}"')
        else:
            processed_query.append(query[i])
        i += 1
    
    query = "".join(processed_query)
    filters = {}
    for part in query.split():
        if ":" in part:
            field, value = part.split(":", 1)
            filters[field.lower()] = value.replace("_", " ").lower()
    return filters

def search_movies(filters):
    ranked = []
    for movie in movies:
        score = 0
        for key, value in filters.items():
            if key in movie:
                if value.lower() in str(movie[key]).lower():
                    score += 1
        if score > 0:
            ranked.append((score, movie))
    
    ranked.sort(reverse=True, key=lambda x: x[0])
    return ranked

query = input("Enter your search terms (e.g. director:\"James Cameron\" genre:Drama year:1994): ")
filters = process_query(query)
ranked_movies = search_movies(filters)

for score, movie in ranked_movies:
    print(f"{movie['title']}, Director: {movie['director']}, Year: {movie['year']}, Genre: {movie['genre']}")
```

```python
import csv

def load_movies(filename):
    movies = []
    with open(filename, 'r', encoding='utf-8') as file:
        csv_reader = csv.DictReader(file)
        for row in csv_reader:
            row['genre'] = row['genre'].replace('|', ',').strip()
            row['year'] = int(row['year'])
            movies.append(row)
    return movies
```

```python
import functools
from fuzzywuzzy import process

movies = [
    {"title": "The Shawshank Redemption", "director": "Frank Darabont", "year": 1994, "genre": "Drama"},
    {"title": "The Godfather", "director": "Francis Ford Coppola", "year": 1972, "genre": "Crime, Drama"},
    {"title": "The Dark Knight", "director": "Christopher Nolan", "year": 2008, "genre": "Action, Crime, Drama"},
    {"title": "Forrest Gump", "director": "Robert Zemeckis", "year": 1994, "genre": "Drama, Romance"},
    {"title": "Inception", "director": "Christopher Nolan", "year": 2010, "genre": "Action, Adventure, Sci-Fi"},
    {"title": "The Matrix", "director": "Lana Wachowski, Lilly Wachowski", "year": 1999, "genre": "Action, Sci-Fi"},
    {"title": "Avengers: Infinity War", "director": "Anthony Russo, Joe Russo", "year": 2018, "genre": "Action, Adventure, Sci-Fi"},
    {"title": "Back to the Future", "director": "Robert Zemeckis", "year": 1985, "genre": "Adventure, Comedy, Sci-Fi"},
    {"title": "The Lion King", "director": "Roger Allers, Rob Minkoff", "year": 1994, "genre": "Animation, Adventure, Drama"},
    {"title": "Pulp Fiction", "director": "Quentin Tarantino", "year": 1994, "genre": "Crime, Drama"},
    {"title": "Fight Club", "director": "David Fincher", "year": 1999, "genre": "Drama"},
    {"title": "Interstellar", "director": "Christopher Nolan", "year": 2014, "genre": "Adventure, Drama, Sci-Fi"},
    {"title": "Spirited Away", "director": "Hayao Miyazaki", "year": 2001, "genre": "Animation, Adventure, Family"},
    {"title": "La La Land", "director": "Damien Chazelle", "year": 2016, "genre": "Comedy, Drama, Music"},
    {"title": "Jurassic Park", "director": "Steven Spielberg", "year": 1993, "genre": "Action, Adventure, Sci-Fi"},
    {"title": "Titanic", "director": "James Cameron", "year": 1997, "genre": "Drama, Romance"},
    {"title": "The Lord of the Rings: The Fellowship of the Ring", "director": "Peter Jackson", "year": 2001, "genre": "Adventure, Drama, Fantasy"},
    {"title": "Star Wars: Episode IV - A New Hope", "director": "George Lucas", "year": 1977, "genre": "Action, Adventure, Fantasy"},
    {"title": "Goodfellas", "director": "Martin Scorsese", "year": 1990, "genre": "Biography, Crime, Drama"},
    {"title": "The Silence of the Lambs", "director": "Jonathan Demme", "year": 1991, "genre": "Crime, Drama, Thriller"},
]

@functools.lru_cache(maxsize=None)
def search_movies(query):
    filters = process_query(query)
    ranked = []
    for movie in movies:
        score = 0
        for key, value in filters.items():
            if key in movie:
                # Exact match
                if value in str(movie[key]).lower():
                    score += 1
                # Fuzzy match
                else:
                    best_match = process.extractOne(value, [str(movie[key]).lower()])
                    if best_match and best_match[1] > 80:
                        score += 1
        if score > 0:
            ranked.append((score, movie))

    ranked.sort(reverse=True, key=lambda x: x[0])
    return ranked

def process_query(query):
    processed_query = []
    i = 0
    while i < len(query):
        if query[i] == '"':
            i += 1
            start = i
            while i < len(query) and query[i] != '"':
                i += 1
            if i < len(query):
                quoted = query[start:i]
                quoted = quoted.replace(" ", "_")
                processed_query.append(f'"{quoted}"')
        else:
            processed_query.append(query[i])
        i += 1

    query = "".join(processed_query)
    filters = {}
    for part in query.split():
        if ":" in part:
            field, value = part.split(":", 1)
            filters[field.lower()] = value.replace("_", " ").lower()
    return filters

query = input("Enter your search terms (e.g. director:\"James Cameron\" genre:Drama year:1994): ")
ranked_movies = search_movies(query)

if ranked_movies:
    for score, movie in ranked_movies:
        print(f"{movie['title']}, Director: {movie['director']}, Year: {movie['year']}, Genre: {movie['genre']}")
else:
    print("No results found.")
```
